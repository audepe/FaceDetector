```python
import urllib.request

def store_image(url, local_file_name):
  with urllib.request.urlopen(url) as resource:
    with open(local_file_name, 'wb') as f:
      f.write(resource.read())
```

```python
store_image('https://img-cdn.hipertextual.com/files/2020/02/hipertextual-hbo-hace-oficial-regreso-friends-2020378966.jpg?strip=all&lossy=1&quality=65&resize=740%2C490&ssl=1', 'test.jpg')
```

```python
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
from mtcnn.mtcnn import MTCNN

image = plt.imread('test.jpg')
plt.imshow(image)

detector = MTCNN()
faces = detector.detect_faces(image)
for face in faces:
    print(face)
```

```python
def highlight_faces(image_path, faces):
  # display image
    image = plt.imread(image_path)
    plt.imshow(image)

    ax = plt.gca()
    # for each face, draw a rectangle based on coordinates
    for face in faces:
        x, y, width, height = face['box']
        face_border = Rectangle((x, y), width, height,
                          fill=False, color='red')
        ax.add_patch(face_border)
    plt.show()
```

```python
highlight_faces('test.jpg',faces)
```
